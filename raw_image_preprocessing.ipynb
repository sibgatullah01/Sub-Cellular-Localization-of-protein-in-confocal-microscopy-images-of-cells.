```python
main_dir = "D:\\ahmad"
def load_data(dir_path):
    images = []
    images_ids = []
    t1 = time()
    for folders in os.listdir(main_dir):
        if folders == "train":
            train_dir = os.path.join(main_dir,folders)
            for image in tqdm(os.listdir(train_dir)):
                image_path = os.path.join(train_dir,image)
                img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE )
                img = cv2.resize(img, (640,640))
                #img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                #img = img/255
                images.append(img)
                images_ids.append(image)
        if (folders.endswith('.csv')):
                labels = pd.read_csv(os.path.join(main_dir,folders))
    t2 = time()
    print(f'Time taken : {(t2-t1)/60}')
    return images, images_ids, labels
```

```python
(images, image_ids, labels) = load_data(main_dir)
```

```python
def copy_file(source_dir, des_dir, img_list):
    import shutil
    for i in img_list:
        src = os.path.join(source_dir,i)
        dst = os.path.join(des_dir,i)
        shutil.copyfile(src, dst)
```
